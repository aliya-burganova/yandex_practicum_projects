## Задача

Провести исследовательский анализ данных Яндекс.Недвижимости о продажах квартир в Санкт-Петербурге и соседних населенных пунктов за несколько лет. Часть данных вписана польхователями вручную, другая часть данных получена автоматически на основе картографических данных.

- `airports_nearest` — расстояние до ближайшего аэропорта в метрах (м)

- `kitchen_area` — площадь кухни в квадратных метрах (м²)
- `last_price` — цена на момент снятия с публикации

- `parks_around3000` — число парков в радиусе 3 км
- `parks_nearest` — расстояние до ближайшего парка (м)
- `ponds_around3000` — число водоёмов в радиусе 3 км

- `rooms` — число комнат
- `studio` — квартира-студия (булев тип)

### Работа с названиями населенных пунктов

Перед тем как приступить к обработке пропусков, приведем названия населенных пунктов в порядок.

Проверим есть ли пропуски в столбце `locality_name`.

```python
data['locality_name'].isna().sum()
```

49 пропусков. Можно заменить эти пропуски некоторой строкой, например `unknown`, однако в дальнейшем названия населенных пунктов мы будем использовать, чтобы на их основе заполнять пропуски в других столбцах и `unknown` может нам подпортить картину.<br><br>
Пропуски в `locality_name` составляют 0.2% от всех данных. Удалим эти строки.

```python
data = (
    data
    .drop(
        data[data['locality_name'].isna()]
        .index
    )
    .reset_index(drop=True)
) 
```

Посмотрим, сколько сейчас уникальных названий населенных пунктов в столбце `locality_name`.

```python
data['locality_name'].nunique()
```

Выведем отсортированный список уникальных названий.

```python
sorted(data['locality_name'].unique())
```

Заметим, что все названия начинаются с большой буквы, в то время как приписки к названиям - с маленькой.
Избавимся от различных приписок, таких как `городской поселок`, `поселок городского типа`, `посёлок`. Названия запишем в новый столбец `locality_name_only`.

```python
data['locality_name_only'] = (
    data['locality_name']
    .str
    .extract('([А-Я].*)')
)
```

Также заменим буквы `ё` на `е`.

```python
data['locality_name_only'] = (
    data['locality_name_only']
    .apply(lambda name: name.replace('ё', 'е'))
)
```

Проверим сколько уникальных названий в столбце `locality_name_only`.

```python
data['locality_name_only'].nunique()
```

```python
sorted(data['locality_name_only'].unique())
```

В результате обработки столбца `locality_name` нам удалось избавиться от 59 неявных дубликатов в названиях населенных пунктов.

### Обработка пропусков

Проверим общее число пропусков в каждом столбце датафрейма

```python
data.isna().sum()
```

Для начала добавим еще один столбец в датафрейм с категорией цены на квартиру.

```python
data['last_price'].describe().round(0)
```

```python
q_25 = 3_400_000
q_50 = 4_650_000
q_75 = 6_799_000
```

```python
def categorize_price(price):
    if price <= q_25:
        return 'низкая'
    if price <= q_50:
        return 'ниже среднего'
    if price <= q_75:
        return 'выше среднего'
    return 'высокая'
```

```python
data['price_category'] = data['last_price'].apply(categorize_price)
```

#### Балконы, апартаменты

Пропуски в столбце `balcony` заполним нулем, так как скорее всего, если пользователь не указал количество балконов, то их нет.

```python
data['balcony'] = data['balcony'].fillna(0)
```

Пропуски в столбце `is_apartement` предположительно появились из-за того, что пользователь не знает разницы между квартирой и апартаментами, а значит у него скорее всего не апартаменты. Заполним пропуски значением `False`.

```python
data['is_apartment'] = data['is_apartment'].fillna(False)
```

#### Этажность

Заполним пропуски в столбце `floors_total` наиболее часто встречающимся значением этажности домов в каждом населенном пункте. При этом, если этаж квартиры окажется больше чем этажность дома, то этажность мы заменим значением этажа квартиры.

```python
data['floors_total'] = (
    data['floors_total'].
    fillna(
        data
        .groupby('locality_name_only')['floors_total']
        .transform(lambda x: x.mode()[0])
    )
)
```

```python
data['floors_total'] =(
    data['floors_total']
    .where(
        data['floors_total'] > data['floor'],
        data['floor']
    )
) 
```

```python
((data['floors_total'] - data['floor']) < 0).sum()
```

#### Высота потолков

Проверим среднюю и медианную высоту потолков

```python
data['ceiling_height'].mean()
```

```python
data['ceiling_height'].median()
```

Обычно высота потолков варьируется от 2.5 до 3 метров. Цена квартиры может зависеть и от высоты потолков. Заполним пропуски в столбце `ceiling_height` медианным значеним высоты потолков среди разных ценовых категорий.

```python
data['ceiling_height'] = (
    data['ceiling_height']
    .fillna(
        data
        .groupby('price_category')['ceiling_height']
        .transform('median')
    )
)
```

#### Жилая площадь и кухня

Пропуски в столбцах `living_area` и `kitchen_area` при отсутствии пропусков в столбце `total_area` могли появить от того, что общая площадь была указана в документах на квартиру, а жилая площадь и кухня - нет, самостоятельно пользователь не решился указать.

Вычислим медиану отношения жилой площади к общей площади, сохраним в переменную `living_area_coeff`.

```python
living_area_coeff = (
    (
        data['living_area']
        / data['total_area']
    )
    .median()
) 
living_area_coeff
```

Вычислим медиану отношения площади кухни к общей площади, сохраним в переменную `kitchen_area_coeff`.

```python
kitchen_area_coeff = (
    (
        data['kitchen_area']
        / data['total_area']
    )
    .median()
) 
kitchen_area_coeff
```

Заполним пропуски в столбцах `living_area` и `kitchen_area` значение столбца `total_area` умноженного на соответствующий коэффициент.

```python
data['living_area'] = (
    data['living_area']
    .fillna(
        data['total_area'] 
        * living_area_coeff
    )
)
```

```python
data['kitchen_area'] = (
    data['kitchen_area']
    .fillna(
        data['total_area'] 
        * kitchen_area_coeff
    )
) 
```

#### Расстояние до центра

Пропуски в расстоянии до центра предположительно обусловлены тем, что пользователю предлагалось поставить метку на карте, для автоматического подсчета расстояния, но не все пользователи это сделали.

Рассмотрим сначала расстояние до центра, если квартира находится в Санкт-Петербурге.

```python
data.query('locality_name_only == "Санкт-Петербург"')['cityCenters_nearest'].isna().sum()
```

Удалим 61 пропуск, поскольку без точного адреса мы не восстановим расстояние.

```python
data = (
    data
    .drop(
        data[(data['locality_name_only'] == 'Санкт-Петербург') & (data['cityCenters_nearest'].isna())]
        .index
    )
    .reset_index(drop=True)
)
```

Проверим общее число пропусков в `cityCenters_nearest`.

```python
data['cityCenters_nearest'].isna().sum()
```

Заполним пропуски в расстояниях до центра медианным значением по каждому населенному пункту.

```python
data['cityCenters_nearest'] = (
    data['cityCenters_nearest']
    .fillna(
        data
        .groupby('locality_name_only')['cityCenters_nearest']
        .transform('median')
    )
)
```

Посмотрим как изменилось количество пропусков.

```python
data['cityCenters_nearest'].isna().sum()
```

Удалось избавиться от 621 пропуска. 

Посмотрим на первые 10 населенных пунктов по количеству объявлений, в которых нет пропусков в расстоянии до центра.

```python
data[data['cityCenters_nearest'].notna()]['locality_name_only'].value_counts().head(10)
```

Посмотрим на первые 10 населенных пунктов по количеству объявлений, в которых отсутствуют расстояния до центра.
